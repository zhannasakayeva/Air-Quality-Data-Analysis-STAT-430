```python
# Data Modeling

modeldf = subdf.sort_index(ascending = True)
modeldf = modeldf[['AQI_O3_1hr', 'AQI_O3_4hr', 'AQI_PM10',	'AQI_PM2.5']]
```

```python
modeldf.info()
```

```python
submodeldf = modeldf[modeldf['AQI_PM2.5'].notna()]
```

```python
submodeldf.info()
```

```python
finaldf = submodeldf.dropna()
```

```python
finaldf.shape
```

```python
newtime = finaldf.index + pd.DateOffset(hours = 1)
```

```python
length = finaldf.shape[0]
newmodeldf = finaldf.copy()
finaldf['PM2.5_hr'] = finaldf['AQI_PM2.5']
```

```python
for i in range(length-1):
  if finaldf.index[i+1] == newtime[i]:
    finaldf['PM2.5_hr'][i] = finaldf['AQI_PM2.5'][i+1]
  else:
    finaldf['PM2.5_hr'][i] = None
```

```python
finaldf.info()
```

```python
finalmodeldf = finaldf.dropna()
```

```python
X = finalmodeldf[['AQI_O3_1hr', 'AQI_O3_4hr', 'AQI_PM10']]
y = finalmodeldf['PM2.5_hr']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=1, 
                                   test_size=0.25, 
                                   shuffle=True)
```

```python
import statsmodels.api as sm
X_train = sm.add_constant(X_train)
OLS1 = sm.OLS(y_train, X_train)
OLS1 = OLS1.fit()
```

```python
print(OLS1.summary())
```

```python
X_test = sm.add_constant(X_test)
```

```python
y_pre = OLS1.predict(X_test)
```

```python
y_pred = OLS1.predict(X_train)
```

```python
from statsmodels.tools.eval_measures import rmse
RMSE1 = rmse(y_test, y_pre)
```

```python
y_train - y_pred
```

```python
OLS1.resid.mean()
```

```python
def check_assumption(model, y):
    fitted_vals = model.predict()
    resids = model.resid
    resids_standardized = model.get_influence().resid_studentized_internal

    fig, ax = plt.subplots(1, 2)

    sns.regplot(x=fitted_vals, y=resids, lowess=True, ax=ax[0], line_kws={'color': 'red'})
    ax[0].set_title('Residuals vs Fitted', fontsize=16)
    ax[0].set(xlabel='Fitted Values', ylabel='Residuals')

    sns.regplot(x=fitted_vals, y=y, lowess=True, ax=ax[1], line_kws={'color': 'red'})
    ax[1].set_title('Observed vs. Predicted Values', fontsize=16)
    ax[1].set(xlabel='Predicted', ylabel='Observed')
```

```python
check_assumption(OLS2, y_train)
```

```python
def category_PM(num):
  if num <= 50:
    category = "Good"
  elif num <= 100:
    category = "Moderate"
  elif num <= 150:
    category = "Unhealthy for Sensitive Group"
  else:
    category = "Unhealthy"
  return category
```

```python
length = finalmodeldf.shape[0]
finalmodeldf['Category_hr'] = 'new'
for i in range(length):
  finalmodeldf['Category_hr'][i] = category_PM(finalmodeldf['PM2.5_hr'][i])
```

```python
X = finalmodeldf[['AQI_O3_1hr', 'AQI_O3_4hr', 'AQI_PM10', 'AQI_PM2.5']]
y = finalmodeldf['Category_hr']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=1, 
                                   test_size=0.25, 
                                   shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train, y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)
```

```python
y_pred =rf.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
